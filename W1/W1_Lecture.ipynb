Simple Example of matplotlib.pyplot

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 1000)
y = np.power(x, 2)

plt.plot(x, y)
plt.show()
```

What About Adding Titles and Labels?

```python
f, ax = plt.subplots(1, 1, figsize = (5, 4))

x = np.linspace(0, 10, 1000)
y = np.power(x, 2)
ax.plot(x, y)
ax.set_xlim((1, 5))
ax.set_ylim((0, 30))
ax.set_xlabel('My x label')
ax.set_ylabel('My y label')
ax.set_title('Plot title, including $\Omega$')

plt.tight_layout()
plt.show()
```

Histogram

```python
data = np.random.randn(1000)
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (6, 3))
ax1.hist(data, bins = 30, density = True, color = 'b')
ax2.hist(data, bins = 30, density = True, color = 'r', cumulative = True)

plt.show()
```

Box Plot

```python
samp1 = np.random.normal(loc = 0., scale = 1., size = 100)
samp2 = np.random.normal(loc = 1., scale = 2., size = 100)
samp3 = np.random.normal(loc = 0.3, scale = 1.2, size = 100)
f, ax = plt.subplots(1, 1, figsize = (5, 4))
ax.boxplot((samp1, samp2, samp3))
ax.set_xticklabels(['Sample 1', 'Sample 2', 'Sample 3'])

plt.show()
```
